```python
# Go to the Job Tab
driver.find_element_by_xpath('/html/body/div[6]/header/div/nav/ul/li[3]/a').click()
```

```python
# find the keywords/location search bars:
search_keywords = driver.find_element_by_xpath('/html/body/div[6]/header/div/div/div/div[2]/div[2]/div/div/input[1]')
keywords = "Data Scientist"
search_keywords.send_keys(keywords)    

search_location = driver.find_element_by_xpath('/html/body/div[6]/header/div/div/div/div[2]/div[2]/div/div/input[1]')
location = "Yerevan, Armenia"
search_location.send_keys(location)   
time.sleep(3)
search_location.send_keys(Keys.ENTER)
```

```python
# Extract the urls of Accouncements
link = [""]

for j in range(1, 3):
    driver.find_element_by_xpath('/html/body/div[6]/div[3]/div[3]/div/div/section[1]/div/div/section/div/ul/li['+str(j)+']/button').click()
    
    for i in range(1, 50):
        try:
            link0 = driver.find_element_by_xpath('/html/body/div[6]/div[3]/div[3]/div/div/section[1]/div/div/ul/li['+str(i)+']/div/div/div[1]/div[2]/div[1]/a').get_attribute('href')
            link.append(str(link0))
            time.sleep(2)
            print(link0)
        except:
            pass
```

```python
len(link)
```

```python
data = pd.DataFrame({'jobtitle':[''], 'company':[''], 'jobtype':[''], 'size':[''], 'details':['']})
for i in link[1:]:
    driver.get(i)
    jobtitle = driver.find_element_by_xpath('/html/body/div[6]/div[3]/div/div[1]/div[1]/div/div[1]/div/div[2]/h1').text
    company = driver.find_element_by_xpath('/html/body/div[6]/div[3]/div/div[1]/div[1]/div/div[1]/div/div[2]/div[1]/span[1]/span[1]/a').text
    jobtype = driver.find_element_by_xpath('/html/body/div[6]/div[3]/div/div[1]/div[1]/div/div[1]/div/div[2]/div[2]/ul/li[1]/span').text
    size = driver.find_element_by_xpath('/html/body/div[6]/div[3]/div/div[1]/div[1]/div/div[1]/div/div[2]/div[2]/ul/li[2]/span').text
    details = driver.find_element_by_xpath('/html/body/div[6]/div[3]/div/div[1]/div[1]/div/div[2]/article/div/div[1]').text.split('\n\n')
    
    data = data.append({'jobtitle':jobtitle, 'company':company, 'jobtype':jobtype, 'size':size, 'details':details}, 
                      ignore_index = True)
```
